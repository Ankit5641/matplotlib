```python
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5,6,7]
likes=[190,210,334,377,150,197,255]
```

```python
plt.title("My facebook likes")
plt.xlabel("Number of days ")
plt.ylabel("Number of likes")
plt.plot(days,likes,color="red",marker="o",mfc="blue",mec="yellow")
plt.grid(color='green')
# plt.xlim(3,5)
# plt.ylim(300,500)
# mfc = marker face color
# mec = marker edge color
# mew = marker egde width
# ms = marker size
# ls = line style 
# lw = line width
```

comparing different social media using matplotlib

```python
days = [1,2,3,4,5,6,7]
```

```python
days
```

```python
fb = likes.copy()
insta = [210,305,667,770,324,678,440]
ytube = [410,550,371,478,103,976,111]
```

```python
fb
```

```python
insta
```

```python
ytube
```

```python
plt.title("Social Media Comparision")
plt.xlabel("Number of days")
plt.ylabel("Number of likes")
plt.plot(days,fb,marker='o',label="fbook",color="blue",mfc="yellow")
plt.plot(days,insta,marker='s',label="insta",color="orange",mfc="blue")
plt.plot(days,ytube,marker='p',label="ytube",color="red",mfc="black")
plt.legend()
plt.grid(color='green')
plt.savefig("mygraph.png")
plt.show()
```
